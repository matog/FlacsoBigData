```python
import pandas as pd
import plotly.express as px
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

# Georreferenciación 

## Ejemplo 1 

```python
data = 'data/gis/prestadores_CB3_DATA_STUDIO_QGIS_IV.csv'
df = pd.read_csv(data)
df.head()
```

```python
# two-line code
fig = px.scatter_mapbox(df,
                        lat="DirGeoY",
                        lon="DirGeoX",
                        color="descripcion",
                        zoom=3,
                        mapbox_style='open-street-map')
fig.show()
```

## Ejemplo 2 

```python
# get data
url = 'https://raw.githubusercontent.com/kefeimo/DataScienceBlog/master/2.geo_plot/df_mapbox_demo.csv'
df = pd.read_csv(url)
df.head()
```

```python
# two-line code
fig = px.scatter_mapbox(df,
                        lat="latitude",
                        lon="longitude",
                        color="gender",
                        zoom=3,
                        mapbox_style='open-street-map')
fig.show()
```
